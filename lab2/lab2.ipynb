```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 6)

from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

### Робота з даними

URL: https://archive.ics.uci.edu/ml/datasets/Heart+Disease

Данi: processed.clivelend.data, heart-disease.names

Вiдгук: Остання колонка

Факторна змiнна: Колонка 2

```python
data = pd.read_csv("processed.cleveland.data",
                 names=["age", "sex", "cp", "trestbps", "chol", 
                        "fbs", "restecg", "thalach", "exang", "oldpeak", 
                        "slope", "ca", "thal", "num"])
data
```

```python
data.isna().any()
```

Якщо уважно подивитися на данні, то можна побачити, що вони насправді мають пропуски в колонках "thal", "ca" у форматі "?", можливо в деяких інших також.

```python
data[87:88]
```

```python
data.shape
```

```python
# виправляємо це
df = data[data != "?"]
df = df.dropna()

df.shape
```

Насправді, в цих даних є ще одна проблему, яку зовсім не легко побачити:

```python
print(df["thal"][0])
print(type(df["thal"][0]))
```

Тобто у стовпчиках, звідки ми видаляли знаки запитання, python вважає дані в них рядками, а не числами. 
А більшість наступних функцій бачать рядки, не можуть їх інтерплітувати, але нічого не кажуть, а викидають всі стовпчики з цими даними. Тому цю помилку я знайшов наприкінці всієї роботи, коли дерево відмовляло генеруватися.

```python
df["thal"] = df["thal"].astype(float)
df["ca"] = df["ca"].astype(float)

print(df["thal"][0])
print(type(df["thal"][0]))
```

```python
y = df["num"]
X = df.drop(columns=["num"])
```

### Класифiкатори

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27052022)
```

```python
# https://scikit-learn.org/stable/modules/naive_bayes.html
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)

print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
#print("Неправильно класифікованих із %d точок : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(6)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()

model = KNeighborsClassifier(10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()

model = KNeighborsClassifier(15)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()
```

Бачимо, що вийшло незначно покращити якість моделі, але цей результат не можна назвати хорошим. 

Також бачимо, що ця модель не має великого потенціалу, бо на тренувальній вибірці вдалося отримати точність близько 0.55

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()
```

```python
from sklearn.svm import SVC

model = SVC(kernel='poly')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()

model = SVC(kernel='linear')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()

model = SVC(kernel='rbf')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()
```

Бачимо, що за допомогою цих моделей не вдається покращити результат. Вони також не мають набагато більшого потенціалу на тренувальній вибірці. 

```python
import xgboost
model = xgboost.XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()
```

```python
model = xgboost.XGBClassifier(max_depth=4)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()
```

```python
model = RandomForestClassifier(max_depth=8)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_1 = model.predict(X_train)
print(f"Точність моделі на тренувальній вибірці = {accuracy_score(y_train, y_pred_1)}")
print(f"Точність моделі на тестовій вибірці= {accuracy_score(y_test, y_pred)}")
print()
```

Дерева дають точність на тренувальній вибірці близько 1 для достатньо великої глибини, але все ж таки не можуть значно покращити результат та перескочити через позначку 0.65 в точності на тестовій вибірці, як і всі інші варіанти до цього. Звичайно, 65% це не так погано для 4х вариіантів класифікації, але звичайно хотілося-б більшого.
